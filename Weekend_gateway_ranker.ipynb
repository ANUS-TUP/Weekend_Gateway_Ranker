```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(r"Top Indian Places to Visit.csv")
df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns.tolist()
```

```python
CITY_COL = "city"
STATE_COL = "state"
LAT_COL = "latitude"       
LON_COL = "longitude"      
RATING_COL = "google_review_rating"
POPULARITY_COL = "number_of_google_review_in_lakhs"
```

```python
df = df.dropna(subset=[CITY_COL, STATE_COL, RATING_COL, POPULARITY_COL])
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
def distance_score_proxy(source_row, target_row):
    if source_row[CITY_COL] == target_row[CITY_COL]:
        return 0
    
    if source_row[STATE_COL] == target_row[STATE_COL]:
        return 1.0   
    
    if source_row["zone"] == target_row["zone"]:
        return 0.6   
    
    return 0.3    
```

```python
def recommend_destinations(source_city, top_n=5):
    source_city = source_city.strip().lower()
    
    if source_city not in df[CITY_COL].str.lower().values:
        return f"Source city '{source_city}' not found in dataset."
    
    source = df[df[CITY_COL].str.lower() == source_city].iloc[0]
    recs = df.copy()
    
    
    recs["distance_score"] = recs.apply(
        lambda row: distance_score_proxy(source, row),
        axis=1
    )
    
    
    recs = recs[recs[CITY_COL].str.lower() != source_city]
    
    
    recs["rating_score"] = recs[RATING_COL] / recs[RATING_COL].max()
    recs["popularity_score"] = recs[POPULARITY_COL] / recs[POPULARITY_COL].max()
    
   
    recs["final_score"] = (
        0.4 * recs["distance_score"] +
        0.4 * recs["rating_score"] +
        0.2 * recs["popularity_score"]
    )
    
    return recs.sort_values(
        by="final_score", ascending=False
    ).head(top_n)[[
        "name", CITY_COL, STATE_COL,
        RATING_COL, POPULARITY_COL, "final_score"
    ]]
```

```python
recommend_destinations("Mumbai")
```

```python
recommend_destinations("Bengaluru")
```

```python
recommend_destinations("Delhi")
```

```python
# Approximate distance matrix between major cities (in KM)
CITY_DISTANCE_MAP = {
    ("Bangalore", "Mysore"): 145,
    ("Bangalore", "Coorg"): 252,
    ("Bangalore", "Ooty"): 270,
    ("Delhi", "Agra"): 233,
    ("Delhi", "Jaipur"): 281,
    ("Mumbai", "Pune"): 149,
    ("Mumbai", "Lonavala"): 83,
}

def get_distance(source, destination):
    if source == destination:
        return 0
    return CITY_DISTANCE_MAP.get(
        (source, destination),
        CITY_DISTANCE_MAP.get((destination, source), 300)  # default distance
    )
```

```python
def add_distance_column(df, source_city):
    df["distance_km"] = df["city"].apply(
        lambda x: get_distance(source_city, x)
    )
    return df
```

```python
def normalize_column(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
def calculate_score(df):
    df["norm_rating"] = normalize_column(df["google_review_rating"])
    df["norm_popularity"] = normalize_column(df["number_of_google_review_in_lakhs"])
    df["norm_distance"] = normalize_column(df["distance_km"])

    df["score"] = (
        df["norm_rating"] * 0.5 +
        df["norm_popularity"] * 0.3 -
        df["norm_distance"] * 0.2
    )
    return df
```

```python
def recommend_weekend_getaways(df, source_city, top_n=5):
    temp_df = df.copy()

    temp_df = add_distance_column(temp_df, source_city)
    temp_df = calculate_score(temp_df)

    recommendations = (
        temp_df[temp_df["city"].str.lower() != source_city.lower()]
        .sort_values("score", ascending=False)
        .head(top_n)
    )

    return recommendations[
        ["city", "state", "distance_km",
         "google_review_rating", "number_of_google_review_in_lakhs", "score"]
    ]
```

```python
recommend_weekend_getaways(df, "Bangalore")
```

```python
recommend_weekend_getaways(df, "Delhi")
```

```python
recommend_weekend_getaways(df, "Mumbai")
```
